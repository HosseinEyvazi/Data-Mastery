
issues with dataframes :
*   we can slice dataframe by their column name only (not with the number)
*   we are able to slice the samples !
\
but we are  unable to : slice the both together by number and slice features by number
\
solution : iloc method


*   loc : label based indexing
*   iloc : index based indexing




```python
# run each of lines

# to slice the df by columns name
df[['name of artist' , 'year']]
# to slice the df by samples index
df[5:8]
df[['name of artist' , 'year']][5:8]
#these desnt work :
#df[5:8 , 'name of artist']
#df[5:8 , 2:4]

#solution : iloc
df.iloc[:8 , :4]
```
